```python
# Libraries to be imported
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import time
```

```python
# options to be used
options = Options()
#options.add_argument('--headless')  
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
```

```python
# Driver of web
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)
```

```python
# URL of main page
url = 'https://www.fresha.com/lp/en/bt/hair-salons/in/ae-dubai'
driver.get(url)
time.sleep(5)
```

```python
# Declaring list to store extracted data
saloon_data = []
```

```python
# Extracting cards by selector
cards = driver.find_elements(By.CSS_SELECTOR, "div.paUULP.Z1aKNU")  
```

```python
# Total cards to be extracted
print(len(cards))
```

```python
# Iterting through all the extracted card whh are 102
for i in range(len(cards)): 
    try:
        # Maintaing the main page by CSS Selector
        cards = driver.find_elements(By.CSS_SELECTOR, "div.paUULP.Z1aKNU")
        card = cards[i]

        # Extract tags data
        p_tags = card.find_elements(By.TAG_NAME, "p")
        span_tags = card.find_elements(By.TAG_NAME, "span")

        # Extracting title, rating and address with respect to tags
        title = p_tags[0].text if len(p_tags) > 0 else "Title not found"
        rating = p_tags[1].text if len(p_tags) > 1 else "Rating not found"
        address = p_tags[3].text if len(p_tags) > 3 else "Address not found"

        # Extracting Dealings
        dealings = "Not specified"
        for span in span_tags:
            text = span.text.lower()
            if "women only" in text:
                dealings = "Women only"
                break
            elif "men only" in text:
                dealings = "Men only"
                break

        # Extracting Services & Costs
        Service_1 = p_tags[4].text if len(p_tags) > 4 else "Service_1 not found"
        Cost_1 = p_tags[6].text if len(p_tags) > 6 else "Cost_1 not found"
        Service_2 = p_tags[7].text if len(p_tags) > 7 else "Service_2 not found"
        Cost_2 = p_tags[9].text if len(p_tags) > 9 else "Cost_2 not found"
        Service_3 = p_tags[10].text if len(p_tags) > 10 else "Service_3 not found"
        Cost_3 = p_tags[12].text if len(p_tags) > 12 else "Cost_3 not found"
        Service_4 = p_tags[13].text if len(p_tags) > 13 else "Service_4 not found"
        Cost_4 = p_tags[15].text if len(p_tags) > 15 else "Cost_4 not found"

        # Clicking "See all services" to move to new page
        try:
            see_all_button = card.find_element(By.CSS_SELECTOR, "p.axDOAG")
            driver.execute_script("arguments[0].click();", see_all_button)
            time.sleep(3)
        except Exception as e:
            print("Could not click on 'See all services':", e)

        # Extract timings after clicking from new page
        monday = tuesday = wednesday = thursday = friday = saturday = sunday = "Not found"
        try:
            # Using tage to extract time of each day
            timing_tags = driver.find_elements(By.TAG_NAME, "p")
            for j in range(len(timing_tags)):
                text = timing_tags[j].text.lower()
                if "monday" in text and j + 1 < len(timing_tags):
                    monday = timing_tags[j + 1].text
                elif "tuesday" in text and j + 1 < len(timing_tags):
                    tuesday = timing_tags[j + 1].text
                elif "wednesday" in text and j + 1 < len(timing_tags):
                    wednesday = timing_tags[j + 1].text
                elif "thursday" in text and j + 1 < len(timing_tags):
                    thursday = timing_tags[j + 1].text
                elif "friday" in text and j + 1 < len(timing_tags):
                    friday = timing_tags[j + 1].text
                elif "saturday" in text and j + 1 < len(timing_tags):
                    saturday = timing_tags[j + 1].text
                elif "sunday" in text and j + 1 < len(timing_tags):
                    sunday = timing_tags[j + 1].text
        except Exception as e:
            print("Error extracting timings:", e)

         # Go back to the main page
        driver.back()
        time.sleep(3)

        
        # Store extracted info
        saloon_data.append({
            "Title": title,
            "Rating": rating,
            "Address": address,
            "Dealings": dealings,
            "Service_1": Service_1,
            "Cost_1": Cost_1,
            "Service_2": Service_2,
            "Cost_2": Cost_2,
            "Service_3": Service_3,
            "Cost_3": Cost_3,
            "Service_4": Service_4,
            "Cost_4": Cost_4,
            "Monday": monday,
            "Tuesday": tuesday,
            "Wednesday": wednesday,
            "Thursday": thursday,
            "Friday": friday,
            "Saturday": saturday,
            "Sunday": sunday
        })
    except Exception as e:
        print(f"Error processing card {i}: {e}")
    if (i == len(cards)-1):
        break
```
